Repeteix l'exercici 1 amb el dataset que disposem en el repositori de GitHub PRE-PROCESSING-DATA, [movies.dat](https://raw.githubusercontent.com/leobioinf0/visualitzacio_grafica_de_multiples_variables/main/movies.dat)

```python
movies = pd.read_csv('https://raw.githubusercontent.com/leobioinf0/visualitzacio_grafica_de_multiples_variables/main/movies.dat',
                     sep = '::',
                     header = None,
                     usecols=[1, 2],
                     names = ['title(year)', 'genre'],
                     encoding="iso-8859-1")
movies.head()
```

### **Una Variable**: 1 Numérica ('Number of genres in a movies')


```python
# We extract the genres
generes = []
for igenre in movies.genre:
    for genere in igenre.split('|'):
        if genere not in generes:
            generes.append(genere)
generes
```

```python
# We create an array of 'zeros', to finally create a DaFrame of 'zeros':

zero_matrix = np.zeros((len(movies), len(generes)))
df = pd.DataFrame(zero_matrix, columns = generes)

# Now, we iterate each movie and set the entries in each row of df to 1. 
# To do this, we use df.columns to calculate the column indices for each genre:
for i, gen in enumerate(movies.genre):
    indices = df.columns.get_indexer(gen.split('|'))
    df.iloc[i, indices] = 1
    df = df.astype('int64')
df.head()
```

```python
# boxplot of the number of genres in a movie
plt.title("Number of genres in a movie")
sns.boxplot(data=df.sum(axis=1),orient='v');
```

### **Dos Variables**: 1 Categórica = ('genre'), 1 Numérica = 'number of movies'

```python
# Sum up the number of movies by genre
n_movies_by_genre = df.sum().sort_values()
n_movies_by_genre
```

```python
# Barplot

ax = sns.barplot(x=n_movies_by_genre.index,
                 y=n_movies_by_genre.values)
ax.set_title("Number of movies by genre")
ax.set_ylabel('Number of movies')
ax.set_xlabel('Genre')
plt.xticks(rotation=90)
plt.show()
```

### **Tres Variables**: 2 Categoricas = ('decade', 'genre'), 1 Numérica = ('Number of movies')

```python
df['decade'] = (pd.to_numeric(movies['title(year)'].str.extract('.*\((.*)\).*')[0])//10)*10
df['decade'] = df["decade"].astype(str)
df.head()
```

```python
# Sum up the number of movies by genre in each decade
df_d = df.groupby(by="decade")[generes].sum()
df_d.head()
```

```python
# create a palette. A color for a genre
colors =  plt.cm.tab20( (4./3*np.arange(len(df.columns)*3/4)).astype(int))

# Barplot
ax = df_d.plot.bar(figsize=(15,7), stacked=True, color=colors)
ax.set_title("Number of movies by decade by genre")
ax.set_ylabel('Number of movies')
ax.set_xlabel('Decade')
plt.show()
```

### **Cuatro Variables**: 3 Categoricas = ('decade', 'n_genres', 'genre'), 1 Numérica = 'number of movies'

```python
df['n_genres'] = df.sum(axis=1)
df_g_d = df.groupby(['n_genres','decade']).sum()
df_g_d.head()
```

```python
# Barplot
fig, axes = plt.subplots(2, 3, figsize=(20, 12))
fig.suptitle('Number of movies by decade (by number of genres and genre)')

df_g_d.loc[(1)].plot.bar(stacked=True,ax=axes[0, 0], legend=None, color=colors)
axes[0, 0].set_title("One genre movies")
axes[0, 0].tick_params(axis='x', rotation=0)

df_g_d.loc[(2)].plot.bar(stacked=True,ax=axes[0, 1], legend=None, color=colors)
axes[0, 1].set_title("Two genres movies")
axes[0, 1].tick_params(axis='x', rotation=0)

df_g_d.loc[(3)].plot.bar(stacked=True,ax=axes[0, 2], color=colors)
axes[0, 2].set_title("Tree genres movies")
axes[0, 2].legend(loc='center left', bbox_to_anchor=(1, 0.5))
axes[0, 2].tick_params(axis='x', rotation=0)

df_g_d.loc[(4)].plot.bar(stacked=True,ax=axes[1, 0], legend=None, color=colors)
axes[1, 0].set_title("Four genres movies")
axes[1, 0].tick_params(axis='x', rotation=0)

df_g_d.loc[(5)].plot.bar(stacked=True,ax=axes[1, 1], legend=None, color=colors)
axes[1, 1].set_title("Five genres movies")
axes[1, 1].tick_params(axis='x', rotation=0)

df_g_d.loc[(6)].plot.bar(stacked=True,ax=axes[1, 2], legend=None, color=colors)
axes[1, 2].set_title("Six genres movies")
axes[1, 2].tick_params(axis='x', rotation=0)
```

### **Tres Variables**: 2 Categoricas = ('decade/year', 'Single genre'), 1 Numérica = ('Number of movies')

```python
# Entries with multiple genres are converted into multiple single genre entries

def attrSplttr(df,col,sep):
    row_accumulator = []
    def splitsToRows(row, sep):
        split_row = row[col].split(sep)
        for s in split_row:
            new_row = row.to_dict()
            new_row[col] = s
            row_accumulator.append(new_row)
    df.apply(splitsToRows, axis=1, args = (sep, ))
    new_df = pd.DataFrame(row_accumulator)
    return new_df

df_s = attrSplttr(df,'genre','|')
df_s
```

```python
plt.figure(figsize=(18,10))
ax = sns.countplot(data=df_s, x="decade", hue="genre")
ax.set_title("Number of movies by decade hued by genre")
plt.legend(loc='upper right')
plt.show()
```

## Second aproach

### [Feature engineering](https://en.wikipedia.org/wiki/Feature_engineering)

The dataset we find the attributes `title (year)` and `genre`. In order to do the following exercises we need to create some new attributes. This objective is achieved through feature engineering.

Feature engineering is the process of using domain knowledge to extract features (characteristics, properties, attributes) from raw data.

```python
# movie_id from object to category
movies['genre'] = movies['genre'].astype("category")

# we extract the year from title(year)
movies['year'] = movies['title(year)'].str.extract('.*\((.*)\).*')
# year from integer to year
movies['year'] = pd.to_numeric(movies['year'])

# decade
movies['decade'] = (movies['year']//10)*10

# lustrum
movies['lustrum'] = movies['year'].apply(lambda x: 'First' if x%10<5 else 'Second')

# period related to second world war
movies['period'] = movies['year'].apply(lambda x: 'Before WWII' if x<1939 else ('During WWII' if x<1945 and x>=1939 else 'After WWII'))

# we extract the title from title(year)
movies['title'] = movies['title(year)'].str.split(r'\(\d*\)', expand=True)[0]
# drop 'title(year)' attribute
movies.drop(columns=['title(year)'], inplace=True)

# number of genres 
movies['n_genre'] = movies['genre'].str.split('|').str.len()

# reorder columns
movies = movies[['title', 'genre', 'n_genre', 'year', 'decade', 'lustrum', 'period']]

movies.head()
```

```python
# Number of titles released per year.
year_df = movies.groupby(by="year").count().sort_index().dropna()[["title"]]
year_df.rename(columns={'title':'n_movies'}, inplace=True)
year_df.head()
```
